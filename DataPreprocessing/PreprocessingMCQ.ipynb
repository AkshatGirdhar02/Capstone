```python
import pandas as pd
```

```python
df = pd.read_csv("../QuestionGenerator/mcqAnswer/questionAnswer.csv")
df
```

### Preprocessing of questions

```python
questions=df['questions']
question=[]
for data in questions:
    entry=data.replace("_","")
    entry=entry.strip().split(".")
    question.append(entry[1])
df['question']=question

```

## Preprocessing of options for mcqs

```python
answers=df['answers']
optionA=[]
optionB=[]
optionC=[]
optionD=[]

for data in answers:
    entry=data.replace(',',"")
    entry=data.replace(" ","")
    entry=entry.split("' '")
    option=entry[0]
    option=option.strip().split('\\n')
    optionA.append(option[0][4:])        
    optionB.append(option[1][2:])        
    optionC.append(option[2][2:])        
    optionD.append(option[3][2:-13])        
    
df['A']=optionA
df['B']=optionB
df['C']=optionC
df['D']=optionD




df
```

```python
answer=[]
explanation=[]
for data in answers:
    entry=data.replace(',',"")
    entry=data.replace(" ","")
    entry=entry.split("' '")
    option=entry[0]
    option=option.strip().split('\\n')
    ans_string=option[4][0:-2]
    x=ans_string.strip().split("Explanation:")
    answer.append(x[0][4:])
    explanation.append(x[1])
df['answer']=answer
df['explanation']=explanation
df
```

```python
df = df.drop(columns=['questions','answers'])
```
